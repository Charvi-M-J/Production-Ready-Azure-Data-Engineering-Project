```python
df_sal = spark.read.table("Bronze.dbo.sales")
```

```python
display(df_sal)
```

```python
df_reve = df_sal.withColumn('RevPerUnit',col('Revenue')/col('Units_Sold'))
display(df_reve)
```

```python
df = df_reve.groupBy('year','BranchName').agg(sum('Units_Sold').alias('Total_Units')).sort('Year','Total_Units',ascending=[1,0])
display(df)
```

```python
df_reve.write.format('parquet')\
         .mode('overwrite')\
         .option('path','abfss://Azure_project@onelake.dfs.fabric.microsoft.com/SilverLH.Lakehouse/Files/carsales')\
         .save()
```
